## Linear Regression Messings
Purpose: Perform routine linear regression using scikit-learn.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lin_inc = np.linspace(0,100,num=100)
test_d = 20*np.random.rand(100)
d_sum = np.add(lin_inc,test_d)
```

```python
#visualize random data
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot( lin_inc,d_sum )
plt.show()
```

```python
#fit the data
x=lin_inc
y=d_sum
reg = LinearRegression()
# Train the model using the training sets
reg.fit( x.reshape(-1,1),y.reshape(-1,1) )
```

```python
#plot data along with regression line
y_predict = reg.predict( x.reshape(-1,1) )
plt.plot( x,y )
plt.plot( x,y_predict,color='black', linewidth=3)
plt.show()
```

```python
# Make predictions using the testing set
x_test=np.linspace(101,200,num=100).reshape(-1,1)
y_predict = reg.predict(x_test)
```
